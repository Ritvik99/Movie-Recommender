```python
n_users = users.user_id.nunique()
n_movies = movies['movie id'].nunique()
```

```python
data_matrix = np.zeros((n_users, n_movies))
for i in ratings.itertuples():
    data_matrix[i[1]-1, i[2]-1] = i[3]
```

```python
data_matrix
```

```python
from sklearn.metrics.pairwise import pairwise_distances

users_similarity = pairwise_distances(data_matrix, metric = 'cosine')
movies_similarity = pairwise_distances(data_matrix.T, metric = 'cosine')
```

```python
def predict(ratings, similarity, type = 'user'):
    
    if type == 'user':
        mean_user_rating = ratings.mean(axis = 1)
        ratings_diff = (ratings - mean_user_rating[:, np.newaxis])
        pred = mean_user_rating[:, np.newaxis] + similarity.dot(ratings_diff)/np.array([np.abs(similarity).sum(axis = 1)]).T
    elif type == 'movies':
        pred = ratings.dot(similarity)/np.array([np.abs(similarity).sum(axis = 1)])
        
    return pred
```

```python
users_prediction = predict(data_matrix, users_similarity)
movies_prediction = predict(data_matrix, movies_similarity, type = 'movies')
```
